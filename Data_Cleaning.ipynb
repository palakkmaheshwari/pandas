```python
# function to get unique values of all the columns in a dataframe

def titanic_unique():
    uni = []
    for i in titanic.columns:
        if i not in ['Age', 'Fare']:
            if titanic[i].dtype == 'O':
                uni.append(f"{i}= {titanic[i].unique()}")
    return uni
        
titanic_unique()
```

```python
# function to get value counts of all the object type

def titanic_valuecounts():
    uni = []
    for i in titanic.columns:
        if i not in ['Age', 'Fare']:
            if titanic[i].dtype == 'O':
                uni.append(f"{i}= {titanic[i].value_counts()}")
    return uni
        
titanic_valuecounts()
```

### Remove Duplicates

`Subset: This parameter specifies the subset of columns to consider when determining duplicates. If specified, only the subset of columns will be used to identify duplicates.`

```python
# find duplicates

titanic.duplicated()
```

```python
# find duplicates values

titanic[titanic.duplicated()]
```

```python
# find the duplicates values and sort then by the columns

titanic[titanic.duplicated()].sort_values(by= list(titanic.columns))
```

```python
# sum of the duplicate values

titanic.duplicated().sum()
```

```python
# find duplicate values

titanic[titanic.duplicated()]
```

```python
# drop the duplicate values row

titanic.drop(list(titanic[titanic.duplicated()].index))
```

```python
# drop duplicate row data

titanic.drop_duplicates(subset=['Survived'])
```

```python
# drop duplicate row data where subset is passed
# subset means the combination of the passed columns (unique)

titanic.drop_duplicates(subset=['Survived', 'Gender'])
```

```python
# drop duplicate row data where subset is passed and keep= last
# subset means the combination of the passed columns (unique)
# keep= last means keep the last duplicate

titanic.drop_duplicates(subset=['Survived', 'Gender'], keep = 'last')
```

```python
# drop duplicate row data where subset is passed, keep= last and ignore_index= True
# subset means the combination of the passed columns (unique)
# keep= last means keep the last duplicate
# ignore_index= True means reset index 

titanic.drop_duplicates(subset=['Survived', 'Gender'], keep = 'last', ignore_index=True)
```
