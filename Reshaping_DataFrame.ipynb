### Pivoting Table
`A pivot table is used to reorganize selected columns and rows of data to summarize the information/insight of the data`<br>
`In short pivot table is reshaping the data`
> **Disadvantage** of pivot is that it do not accept index where there are duplicate values
>> Index containing duplicate entries, cannot reshape

```python
# Read csv file

table1 = pd.read_csv('Pandas_Data\\table1.csv')
table1.head(6)
```

```python
# info

table1.info()
```

```python
# create pivot with column medal

table_medal = table1.pivot(columns= 'Medal')
table_medal
```

```python
# view data columns

table1.pivot(columns= 'Medal').columns
```

```python
# slicing

table_medal.loc[3,('Country',   'Gold')]
```

```python
# slicing

table_medal.iloc[3,1]
```

```python
# slicing

table_medal.iloc[3,4]
```

```python
# create pivot with column medal and value is Count column

table1.pivot(columns= 'Medal', values= 'Count')
```

```python
# create pivot with column medal, value is Count column and index is Country

table1.pivot(index= 'Country', columns= 'Medal', values= 'Count')
```

```python
# create pivot with column medal and index is Country

table1.pivot(index= 'Country', columns= 'Medal')
```

```python
# Fill 0 to all NaN values by fillna

table1_fill = table1.pivot(index= 'Country', columns= 'Medal', values= 'Count').fillna(0)
table1_fill
```

```python
# data info

table1_fill.info()
```

```python
# data head()

table1.head()
```

```python
# set multi index

table1.set_index(['Country', 'Medal'])
```

```python
# unstack() data and fill NaN value to 0

table1.set_index(['Country', 'Medal']).unstack(fill_value= 0)
```

```python
# Read csv file

table2 = pd.read_csv('Pandas_Data\\table2.csv')
table2.head(6)
```

```python
# create pivot with column medal, index country and value count
# limitations of pivot is Index contains duplicate entries, cannot reshape

table2.pivot(index= 'Country', columns= 'Medal', values= 'Count')
```

### pivot_table

```python
# data info

table2.info()
```

```python
# index value_counts

table2.index.value_counts()
```

```python
# data pivot table

table2_country = table2.pivot_table(index= 'Country', columns= 'Medal', values= 'Count', fill_value= 0)
table2_country
```

```python
# index value_counts

table2_country.index.value_counts()
```

```python
# sum of pivot table by aggfunc

table2_medals = table2.pivot_table(index= 'Country', columns= 'Medal', values= 'Count', fill_value= 0, aggfunc= 'sum')
table2_medals
```

```python
# index value_counts

table2_medals.index.value_counts()
```

```python
# create pivot table where index is country, column is medal, value is count, fill NaN values = 0, aggfunc= sum and 
# margin_name= total_medals
# without margin = True, the margins_name will not work

table2.pivot_table(index='Country',columns='Medal',values='Count',fill_value=0,aggfunc='sum',margins_name='Total_Medals')
```

```python
# create pivot table where index is country, column is medal, value is count, fill NaN values = 0, 
# aggfunc= sum and margin_name= total_medals
# If margins_name is not given then the default name will be All

table2.pivot_table(index='Country',columns='Medal',
                   values='Count',fill_value=0,aggfunc='sum',
                   margins=True,margins_name='Total_Medals')
```

```python
# error due to passing more than 1 aggfunc

table2.pivot_table(index='Country',columns='Medal',values='Count',fill_value=0,aggfunc=['sum','mean'],
                   margins=True,margins_name='Total_Medals','Mean_Medals')
```

```python
# group data by country and medal the sum by count and unstack the data

table2_cm_unstack = table2.groupby(['Country', 'Medal'])['Count'].sum().unstack(fill_value= 0)
table2_cm_unstack
```

```python
# add new column name total medals and assign the value

table2_cm_unstack['Total Medals'] = table2_cm_unstack.sum(axis= 1)
table2_cm_unstack
```

```python
# Update the existing column

table2_cm_unstack['Total Medals'] = table2_cm_unstack.sum(axis= 1)
table2_cm_unstack
```

### pd.crosstab()
`With crosstab() we can derived the relationship between 2 or more variables by passing them as index and columns`
> **Parameters**: 
>> 1. index
>> 2. columns
>> 3. margins
>> 4. normalize
>> 5. value
>> 6. aggfunc
>> 7. margins_name
>> 8. fill_value
>> 9. dropna
>> 10. rownames
>> 11. colnames

```python
# info

titanic.info()
```

```python
# find the count of pclass for the values of sex
# using crosstab()

pd.crosstab(index= titanic['sex'], columns= titanic['pclass'])
```

```python
# find the count of pclass for the values of sex, margins = True
# using crosstab()
# margins= True means to give subtotals to all the row and columns

pd.crosstab(index= titanic['sex'], columns= titanic['pclass'], margins= True)
```

```python
# find the count of pclass for the values of sex, margins = True and normalize= True
# using crosstab()
# margins= True means to give subtotals to all the rows and columns by default
# normalize= True means to bring values between 0 and 1

pd.crosstab(index= titanic['sex'], columns= titanic['pclass'], margins= True, normalize= True)
```

```python
# sum of pclass where data is groupby sex and pclass
# convert level 1 index to columns using unstack()

titanic.groupby(['sex','pclass'])['pclass'].sum().unstack()
```

```python
# Read csv file

summer = pd.read_csv('Pandas_Data\\summer.csv')
summer
```

```python
# find the count of pclass for the values of sex,aggfunc= count, values= summer['Athlete'] and margins = True 
# using crosstab()
# aggfunc means it is a aggrigation function which is used to do operation on the columns and rows(like count,sum,mean,etc) 
# values means on the basis of which calculation is done for all rows and columns
# margins= True means to give subtotals to all the rows and columns by default

pd.crosstab(index= [summer['Year'],summer['Country']], 
            columns= summer['Medal'], aggfunc= 'count',
            values= summer['Athlete'], margins= True)
```

```python
# find the count of pclass for the values of sex,aggfunc= count, values= summer['Athlete'] and margins = True 
# using crosstab()
# aggfunc means it is a aggrigation function which is used to do operation on the columns and rows(like count,sum,mean,etc) 
# values means on the basis of which calculation is done for all rows and columns
# margins= True means to give subtotals to all the rows and columns by default or as given in the aggfunc
# fillna = 0 means fill all NaN values to 0

pd.crosstab(index= [summer['Year'],summer['Country']], 
            columns= summer['Medal'], aggfunc= 'count',
            values= summer['Athlete'], margins= True).fillna(0)
```

```python
# create a pivot_table where index= year and country, columns = medals, values= athlete, aggfunc= count and margins= True

summer.pivot_table(index= ['Year','Country'], columns= 'Medal', 
                   values= 'Athlete', fill_value= 0, aggfunc= 'count', margins= True)
```

```python
# value-counts of medals where data is groupby year and country
# convert level 1 index to columns using unstack() 
# fill the NaN(missing values) with 0

summer.groupby(['Year','Country'])['Medal'].value_counts().unstack(fill_value=0)
```

```python
# count of the medal where data is groupby year, country and medal
# convert level 1 index to columns using unstack() 
# fill the NaN(missing values) with 0

summer.groupby(['Year','Country','Medal'])['Medal'].count().unstack(fill_value=0)
```

### melt()
`With melt() we can convert columns as value in a single new column`

> **Parameters:**
>> 1. id_vars
>> 2. value_vars
>> 3. var_name
>> 4. value_name
>> 5. ignore_index

```python
# data describe

table2.describe()
```

```python
# Read csv file

table_2012 = pd.read_csv('Pandas_Data\\table_2012.csv')
table_2012.head()
```

```python
# convert 'value_vars' as a values of 'var_name' as a new column on the basis of 'id_vars' and their value in 'value_name'

table_2012.melt(id_vars= 'Country', value_vars= ['Gold','Silver','Bronze'], var_name= 'Medal', value_name= 'Count')
```

```python
# convert 'value_vars' as a values of 'var_name' as a new column on the basis of 'id_vars' and their value in 'value_name'

table_2012.melt(id_vars= 'Country', var_name= 'Medal', value_name= 'Countt')
```

```python
# convert 'value_vars' as a values of 'var_name' as a new column on the basis of 'id_vars' and their value in 'value_name'

table_2012.melt(id_vars= ['Country', 'Bronze'], value_vars= ['Gold','Silver'], var_name= 'Medal', value_name= 'Count')
```
