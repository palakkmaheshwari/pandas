### concat()

`If 2 dataframes have same columns name and the concat axis is 0 then concat will act as append functionality`

```python
# concat data of men2004 and men2008 (join the data)

pd.concat([men2004, men2008], ignore_index= True)
```

```python
# concat data of men2004 and men2008 (join the data) by resetting the index
# key will not be assigned because ignore_index is True

pd.concat([men2004, men2008], ignore_index= True, keys= [2004, 2008])
```

```python
# concat data of men2004 and men2008 (join the data) by resetting the index
# key and names will not be assigned because ignore_index is True

pd.concat([men2004, men2008], ignore_index= True, keys= [2004, 2008], names= ['Year']) 
```

```python
# concat data of men2004 and men2008 (join the data) by resetting the index
# key will be assigned because ignore_index is False

pd.concat([men2004, men2008], ignore_index= False, keys= [2004, 2008])
```

```python
# concat data of men2004 and men2008 (join the data) by resetting the index
# key and names(Index Name) will be assigned because ignore_index is False

ss = pd.concat([men2004, men2008], ignore_index= False, keys= [2004, 2008], names= ['Year'])
ss
```

```python
ss.reset_index()
```

```python
# reset_index and remove column 'level_1'

ss.reset_index().drop(columns= 'level_1')
```

## Merge

`To merge, 2 dataframes there should be atleast 1 common feature/column name`

### Outer Join

```python
# head

men2004.head(), men2004.shape
```

```python
# head

men2008.head(), men2008.shape
```

```python
# merge data of men2004 with men2008 where how= outer
# outer means merge all the same data of both the dataframe and append rest data from both the dataframe

men2004.merge(men2008, how= 'outer')
```

```python
men2004.merge(men2008, how= 'inner').shape
```

```python
# Check the shape for the outer join

men2004.shape[0]+ men2008.shape[0]- men2004.merge(men2008, how= 'inner').shape[0]
```

```python
# Merge 2 dataframe as outer join on the basis of Athlete column

men2004.merge(men2008, how= 'outer', on= 'Athlete')
```

```python
# Merge 2 dataframe as outer join on the basis of Athlete column and adding suffixes on the new columns

men2004.merge(men2008, how= 'outer', on= 'Athlete', suffixes= ['_2004','_2008'])
```

```python
# Merge 2 dataframe as outer join on the basis of Athlete column, adding suffixes and indicator
# indicator tell about the type of join

men2004.merge(men2008, how= 'outer', on= 'Athlete', suffixes= ['_2004','_2008'], indicator= True)
```

### left & right join

```python
# Merge 2 dataframe as left join on the basis of Athlete column, adding suffixes and indicator
# indicator tell about the type of join
# Left join takes all the data from left and the same data of both dataframe 

men2004.merge(men2008, how= 'left', on= 'Athlete', suffixes= ['_2004','_2008'], indicator= True)
```

```python
# Merge 2 dataframe as right join on the basis of Athlete column, adding suffixes and indicator
# indicator tell about the type of join
# right join takes all the data from right and the same data of both dataframe 

men2004.merge(men2008, how= 'right', on= 'Athlete', suffixes= ['_2004','_2008'], indicator= True)
```

```python
import pandas as pd

df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],
                    'value': [1, 2, 3, 4]})

df2 = pd.DataFrame({'key': ['B', 'D', 'E', 'F'],
                    'value': [5, 6, 7, 8]})

# Inner join on the 'key' column
result_inner = pd.merge(df1, df2, on='key', how='inner')
print("Inner Join:")
print(result_inner)

# Left join on the 'key' column
result_left = pd.merge(df1, df2, on='key', how='left')
print("\nLeft Join:")
print(result_left)

# Right join on the 'key' column
result_right = pd.merge(df1, df2, on='key', how='right')
print("\nRight Join:")
print(result_right)

# outer join on the 'key' column
result_outer = pd.merge(df1, df2, on='key', how='outer')
print("\nOuter Join:")
print(result_outer)
```

### Cross Join

```python
# Merge 2 dataframe as cross join, adding suffixes and indicator
# indicator tell about the type of join
# cross join makes all the possible combinations from both dataframe

men2004.merge(men2008, how= 'cross', suffixes= ['_2004','_2008'], indicator= True)
```

### pd.merge() and join()

```python
# read data

men2004 = pd.read_csv('Pandas_Data\\men2004.csv', index_col= 'Athlete')
men2008 = pd.read_csv('Pandas_Data\\men2008.csv', index_col= 'Athlete')
```

```python
# data head()

men2004.head(3)
```

```python
# data head()

men2008.head(3)
```

```python
# merge 2 dataframes by .merge

m1 = men2004.merge(men2008, how= 'outer', on= 'Athlete', suffixes= ['_2004', '_2008'], indicator= True)
m1
```

```python
# merge 2 dataframes by pd.merge

m2 = pd.merge(men2004, men2008, how= 'outer', on= 'Athlete', suffixes= ['_2004', '_2008'], indicator= True)
m2
```

```python
# check .merge and pd.merge are same or not

m1.equals(m2)
```
